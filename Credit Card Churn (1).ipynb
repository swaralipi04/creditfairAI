```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (assuming 'df' is already defined)
# df = pd.read_csv('data.csv')

# List of numerical columns
numerical_columns = [
    'Customer_Age', 'Dependent_count', 'Months_on_book', 'Months_Inactive_12_mon',
    'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal', 'Avg_Open_To_Buy',
    'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1',
    'Avg_Utilization_Ratio'
]

# Step 1: Identify outliers using IQR method
plt.figure(figsize=(20, 15))
for i, column in enumerate(numerical_columns):
    plt.subplot(3, 5, i + 1)  # Adjusted subplot grid to 3 rows, 5 columns
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column} with Outliers')
    plt.xlabel(column)
plt.tight_layout()
plt.show()

# Step 2: Remove outliers from the dataset
df_cleaned = df.copy()  # Create a copy of the original dataframe

for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Remove outliers
    df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]

# Optionally, compare the original and cleaned dataset sizes
print(f"Original dataset shape: {df.shape}")
print(f"Cleaned dataset shape: {df_cleaned.shape}")

# Step 3: Visualize distribution after removing outliers
plt.figure(figsize=(20, 15))
for i, column in enumerate(numerical_columns):
    plt.subplot(3, 5, i + 1)  # Adjusted subplot grid to 3 rows, 5 columns
    sns.boxplot(x=df_cleaned[column])
    plt.title(f'Boxplot of {column} without Outliers')
    plt.xlabel(column)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
import matplotlib.pyplot as plt
import numpy as np

# Initialize classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Support Vector Machine': SVC(random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'XGBoost': XGBClassifier(random_state=42),
    'CatBoost': CatBoostClassifier(random_state=42, verbose=False)
}

# Lists to store classifier names and their accuracies
clf_names = []
accuracies = []

# Train and evaluate each classifier
for clf_name, clf in classifiers.items():
    clf.fit(X_train_scaled, y_train)
    y_pred = clf.predict(X_test_scaled)
    
    # Store classifier name
    clf_names.append(clf_name)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    
    # Print classifier name and accuracy
    print(f"Classifier: {clf_name}")
    print(classification_report(y_test, y_pred))
    print(f"Accuracy: {accuracy:.2f}\n")

# Plotting the accuracies
plt.figure(figsize=(10, 6))
plt.barh(clf_names, accuracies, color='skyblue')
plt.xlabel('Accuracy')
plt.title('Classifier Accuracies')
plt.xlim(0.7, 1.0)  # Adjust limits based on your accuracy range
plt.gca().invert_yaxis()  # Invert y-axis to show the highest accuracy at the top
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from fairlearn.metrics import demographic_parity_difference, equalized_odds_difference
from fairlearn.reductions import GridSearch, EqualizedOdds

# Assuming df_encoded is your encoded DataFrame

# Split data into features (X) and target (y)
X_encoded = df.drop(columns=['Attrition_Flag'])  # Features
y = df['Attrition_Flag']  # Target

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Standardize numerical features if necessary
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Models to evaluate
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'SVM': SVC(random_state=42)
    
}

# Train and evaluate models
for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train_scaled, y_train)
    
    # Predictions
    y_pred = model.predict(X_test_scaled)
    
    # Accuracy metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    
    # Print metrics
    print(f"\n{model_name} Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")
    print(f"Confusion Matrix:\n{cm}\n")
    
    # Fairness evaluation with Fairlearn
    print(f"Assessing fairness for {model_name}...")
    fairlearn_gridsearch = GridSearch(
        estimator=model,
        constraints=EqualizedOdds(),
        grid_size=71  # Adjust grid size for trade-off between fairness and accuracy
    )
    fairlearn_gridsearch.fit(X_train_scaled, y_train, sensitive_features=X_train['Gender'])
    y_pred_fair = fairlearn_gridsearch.predict(X_test_scaled)
    
    # Fairness metrics
    disparity_gender = demographic_parity_difference(y_true=y_test, y_pred=y_pred_fair, sensitive_features=X_test['Gender'])
    eq_odds_gender = equalized_odds_difference(y_true=y_test, y_pred=y_pred_fair, sensitive_features=X_test['Gender'])
    
    # Print fairness metrics
    print(f"Demographic Parity Difference (Gender): {disparity_gender}")
    print(f"Equalized Odds Difference (Gender): {eq_odds_gender}\n")
```
